```python
paragraph = """Certainly! Expanding on the character of John Wick allows for a deeper exploration of his persona, motivations, and impact within the world he inhabits. From his origins as a retired hitman seeking solace in a quiet life to his transformation into a relentless force of retribution, every aspect of Wick's journey is steeped in nuance and intrigue.

John Wick's enigmatic nature is a central element of his allure. While snippets of his past are revealed throughout the series, much remains shrouded in mystery, leaving ample room for speculation and interpretation. This deliberate ambiguity adds layers to his character, inviting audiences to ponder the events that shaped him into the formidable assassin he has become. Whether it's his rumored ties to a secret society of assassins or the tragic loss of his wife, Helen, Wick's backstory serves as a rich tapestry upon which his actions are woven.

At the heart of John Wick lies a compelling dichotomy between his stoic exterior and the emotional turmoil that simmers beneath the surface. While he may appear as an unflappable killing machine, capable of dispatching adversaries with ruthless efficiency, glimpses of vulnerability emerge in his interactions with those he holds dear. His unwavering devotion to Helen, manifested through his relentless pursuit of justice in her memory, humanizes him amidst the chaos of his world. It's this emotional anchor that grounds Wick, reminding audiences that beneath the facade of a hardened assassin lies a man driven by love and loss.

What truly sets John Wick apart, however, is his unwavering sense of honor and integrity. Despite operating within the morally ambiguous realm of assassins and criminals, Wick adheres to a strict code of conduct that sets him apart from his peers. Whether it's his refusal to harm innocents or his adherence to the rules of engagement within the underworld, Wick operates with a sense of honor that commands respect even from his adversaries. It's this adherence to his own moral compass that elevates him beyond the typical action hero archetype, positioning him as a figure of mythic stature within the world of cinema.

In essence, John Wick transcends the confines of his genre, offering audiences a complex and multifaceted character whose depth and complexity rival that of any dramatic protagonist. From his mysterious origins to his unwavering sense of honor, every aspect of Wick's persona serves to captivate and intrigue, solidifying his place as one of the most iconic characters of the modern era."""
```

```python
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('English')
```

```python
stopwords.words('German')
```

```python
stemmer = SnowballStemmer('english')
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('English'))]
    sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('English'))]
    sentences[i] = ' '.join(words)
```
